```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas dataframe
credit_card_data=pd.read_csv("/content/creditcard.csv")
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['V23'].fillna(credit_card_data['V23'].median(),inplace=True)
credit_card_data['V24'].fillna(credit_card_data['V24'].median(),inplace=True)
credit_card_data['V25'].fillna(credit_card_data['V25'].median(),inplace=True)
credit_card_data['V26'].fillna(credit_card_data['V26'].median(),inplace=True)
credit_card_data['V27'].fillna(credit_card_data['V27'].median(),inplace=True)
credit_card_data['V28'].fillna(credit_card_data['V28'].median(),inplace=True)
credit_card_data['Amount'].fillna(credit_card_data['Amount'].median(),inplace=True)
credit_card_data['Class'].fillna(credit_card_data['Class'].median(),inplace=True)
```

```python
credit_card_data.isnull().sum()
```

```python
#checking legit and fraudulent transactions
#legit will be given as 0
#fraudulent will be given as 1
credit_card_data['Class'].value_counts()
```

```python
#separating the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both the transactions
credit_card_data.groupby('Class').mean()
```

Now we are going to build sample dataset containing similar distribution of normal transactions and fraudulent transactions.
No.of fraudulent transactions-->280

```python
legit_sample=legit.sample(n=280)
```

```python
#concatenating two dataframes
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
#head gives legit samples
new_dataset.head()
```

```python
#tail gives fraud samples
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

Above we can see uniformly distributed legit and fraud transactions

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
                  
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting the data into training data and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train,Y_train)
```

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
training_data_accuracy
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
test_data_accuracy
```
